```python
# Take only min and remove from the original dataframe and re-run the code and add exam date
outMtrx = []
for SID in unique_df:
  sub_diagg_df = sample_df.loc[sample_df['RID'] == SID]
  if len(sub_diagg_df) > 0:
    tmp = [SID, sample_df.loc[sub_diagg_df['EXAMDATE'].idxmin(), 'Diagnosis'], sample_df.loc[sub_diagg_df['EXAMDATE'].idxmin(), 'EXAMDATE']]
    sample_df = sample_df.drop(sub_diagg_df['EXAMDATE'].idxmin())
    outMtrx.append(tmp)
    tmp = []
```
